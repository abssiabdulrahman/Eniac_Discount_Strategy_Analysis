## OUTLIER ANALYSIS

```python
#CALCULATING DISCOUNT PERCENTAGE FROM PRICE AND UNIT_PRICE
orders_products_dt["discount_%"] = (
    (orders_products_dt["price"] - orders_products_dt["unit_price"])
    / orders_products_dt["price"]
) * 100
orders_products_dt["discount_%"].describe().round(2)
```

### DISCOUNT OUTLIERS

```python
#FLAGGING DISCOUNT OUTLIERS
def flag_discount(row):
    if row["unit_price"] < 1:
        return "near-free (unit_price < €1)"
    elif row["discount_%"] > 75:
        return "suspicious (>75% discount)"
    elif row["discount_%"] < 0:
        return "impossible (unit_price > price)"
    else:
        return "normal"

orders_products_dt["discount_flag"] = orders_products_dt.apply(flag_discount, axis=1)
orders_products_dt["discount_flag"].value_counts()
```

```python
#SHOWING THE SUSPICIOUS AND NEAR-FREE ROWS
orders_products_dt.loc[
    orders_products_dt["discount_flag"] != "normal",
    ["order_id", "sku", "name", "unit_price", "price", "discount_%", "discount_flag"]
].sort_values("discount_%")
```

## Data Quality Assessment

A data quality check was performed to review the completeness, validity, and consistency of the dataset before starting the analysis.


## Data Quality Issues

### 1. Pricing inconsistencies (important)
Around 4–5% of the rows contain cases where the unit price is higher than the base price, which should not normally happen.

**Possible reasons:**
- The base price might be outdated  
- Some products (e.g. premium brands) may follow different pricing rules  

**Action taken:**
- Excluded these rows from discount-related analysis  
- Kept them in mind for separate investigation  

---

### 2. Missing values (moderate)
Missing values were mainly found in:
- `price` (important)  
- `desc` and `type` (less important)  

**Impact:**
- Missing prices directly affect discount calculations  

**Action taken:**
- Removed rows where the price was missing or invalid  

---

### 3. Data type issues (resolved)
Some columns were stored incorrectly:
- Prices were stored as text  
- Dates were stored as objects  

**Action taken:**
- Converted `price` and `unit_price` to numeric  
- Converted `created_date` and `date` to datetime  

---

### 4. Total price inconsistencies (moderate)
Some rows show `total_paid` is lower than `unit_price × product_quantity`.

**Possible reasons:**
- Order-level discounts  
- Shipping costs added at order level  
- Vouchers reducing the total  

**Action taken:**
- Kept these rows for analysis  
- Noted as a structural limitation — use `orders_price` for revenue, not `unit_price` alone  

---

### 5. Quantity outliers — B2B bulk orders (moderate)
Most orders have small quantities (1–3 items), but a few rows show very high values (e.g. 32×, 206×, 800×).

**Finding:**
Deep analysis confirmed these are legitimate B2B transactions (schools, businesses, corporate buyers).  
Discount behavior is virtually identical to retail customers (20.9% vs 19.4% avg).  
Combined they represent only 0.35% of total revenue.

**Action taken:**
- Kept in the dataset  
- Flagged as B2B segment for separate interpretation  

---

### 6. Unconfirmed orders — Place Order state (important)
31,965 rows had the state `Place Order`, meaning the transaction was initiated but never confirmed.

**Finding:**
`Place Order` rows show significantly lower average discounts (15.5% vs 21.6%) and higher average unit prices (€474 vs €239) compared to `Completed` orders.  
Including them distorts all metrics.

**Action taken:**
- Dropped all `Place Order` rows  
- Kept only `Completed` transactions  

---

### 7. Zero revenue orders (important)
2 orders had `total_paid = €0`, meaning no revenue was recorded despite products being listed.

**Possible reasons:**
- Test orders  
- Data entry errors  
- Free promotional orders  

**Action taken:**
- Dropped these 2 rows — not legitimate sales  

---

### 8. Near-free service entries (important)
2 rows had `unit_price = €0.01` — RAM installation and SSD installation services sold at near-zero price.

**Finding:**
These are internal service entries, not real product sales. They inflate discount percentages and distort unit price analysis.

**Action taken:**
- Dropped all rows where `unit_price < €1`  

---

### 9. Extreme discount outliers (important)
931 rows had a discount percentage above 75%, reaching up to 99.98%.

**Finding:**
These are concentrated in post-holiday clearance periods (Jan 2017, Oct 2017, Feb 2018) and consist mainly of discontinued brands (Jawbone, Zagg, Griffin).  
They are economically suspicious for a standard tech retailer and skew discount analysis significantly.

**Action taken:**
- Dropped rows where `discount_% > 75%`  

---

### 10. Duplicate order lines (minor)
4 rows were exact duplicates — the same SKU appeared twice within the same order at the same price.

**Possible reasons:**
- Data entry errors  
- System duplication during order processing  

**Action taken:**
- Dropped duplicate rows on `order_id + sku + unit_price`  

---

## Final Data Preparation

To ensure reliable analysis, the following steps were applied in order:

- Filtered to `Completed` orders only — removed 31,965 unconfirmed rows  
- Removed rows with missing or invalid prices  
- Dropped `total_paid = €0` orders — 2 rows  
- Dropped `unit_price < €1` service entries — 2 rows  
- Dropped `discount_% > 75%` clearance outliers — 931 rows  
- Dropped duplicate order-SKU rows — 4 rows  
- Standardized all data types  

---

## Summary

The dataset required significant cleaning, particularly around order state filtering and discount outliers.
